b) Scatter plot with customizations

```python
import matplotlib.pyplot as plt

# Scatter plot between age and smoking
plt.scatter(df['age'], df['smoker'], marker='+')
plt.title('Scatter Plot between Age and Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking (yes/no)')
plt.text(30, 'yes','.', fontsize=12, ha='right')
plt.show()
```

c) Filter the data and update plot

```python
# Scatter plot between age (20 to 40) and smoking
filtered_df = df[(df['age'] >= 20) & (df['age'] <= 40)]
plt.scatter(filtered_df['age'], filtered_df['smoker'], marker='+')
plt.title('Scatter Plot between Age (20 to 40) and Smoking')
plt.xlabel('Age')
plt.ylabel('Smoking')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Select numerical columns (excluding 'expenses')
numerical_cols = df.select_dtypes(include=['number']).columns.difference(['expenses'])

# Create scatter plots for each numerical column against 'expenses'
for col in numerical_cols:
    plt.figure(figsize=(8, 6))  # Adjust figure size as needed
    sns.scatterplot(x=col, y='expenses', data=df)
    plt.title(f'Scatter Plot: {col} vs. Expenses')
    plt.xlabel(col)
    plt.ylabel('Expenses')
    plt.show()
```

```python
# Select categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Create bar plots or box plots for each categorical column against 'expenses'
for col in categorical_cols:
    plt.figure(figsize=(8, 6))  # Adjust figure size as needed
    # Choose either bar plot or box plot based on your preference
    # sns.barplot(x=col, y='expenses', data=df)  # For bar plot
    sns.boxplot(x=col, y='expenses', data=df)  # For box plot
    plt.title(f'{col.capitalize()} vs. Expenses')
    plt.xlabel(col.capitalize())
    plt.ylabel('Expenses')
    plt.show()
```

```python
import pandas as pd

# Assuming your dataset is loaded into a pandas DataFrame called 'df'
# If not, replace 'df' with the actual name of your DataFrame
# df = pd.read_csv('your_dataset.csv')  # Replace 'your_dataset.csv' with the actual file path

# Group by 'sex' and calculate the sum of expenses for each group
gender_expenses = df.groupby('sex')['expenses'].sum()

# Print the results
print(gender_expenses)

# Determine which gender has more expenses
if gender_expenses['male'] > gender_expenses['female']:
    print("Males have more expenses.")
elif gender_expenses['female'] > gender_expenses['male']:
    print("Females have more expenses.")
else:
    print("Both genders have equal expenses.")
```

d) Expenses of smokers and non-smokers

```python
# Find expenses of smokers and non-smokers
smokers_expenses = df[df['smoker'] == 'yes']['expenses']
non_smokers_expenses = df[df['smoker'] == 'no']['expenses']

plt.boxplot([smokers_expenses, non_smokers_expenses], labels=['Smokers', 'Non-Smokers'])
plt.title('Expenses of Smokers vs Non-Smokers')
plt.ylabel('Expenses')
plt.show()
```

e) Number of smokers gender-wise

```python
# Number of smokers gender-wise
smoker_gender_count = df[df['smoker'] == 'yes'].groupby('sex').size()

smoker_gender_count.plot(kind='bar',color = 'brown')
plt.title('Number of Smokers by Gender')
plt.ylabel('Count')
plt.show()
```
